```python
# Import packages

import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the pre-trained mask detection model
model = tf.keras.models.load_model('/Users/akshay/projects/face_mask_model/Netmask_10_epoch.h5')

# Read the sample image
sample_mask_img = cv2.imread('/Users/akshay/projects/face_mask_model/archive/images/maksssksksss62.png')

# Display the sample image
plt.imshow(sample_mask_img)

# Resize and preprocess the image for prediction
sample_mask_img = cv2.resize(sample_mask_img, (128, 128))
sample_mask_img = np.reshape(sample_mask_img, [1, 128, 128, 3])
sample_mask_img = sample_mask_img / 255.0

# Predict using the loaded model
prediction = model.predict(sample_mask_img)

# Print the prediction results
print(f'Accuracy of wearing mask: {prediction[0][0] * 100:.2f}%')
print(f'Accuracy of not wearing mask: {prediction[0][1] * 100:.2f}%')

```
